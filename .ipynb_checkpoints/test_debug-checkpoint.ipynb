```python
import sys
import torch
import importlib

%matplotlib inline

sys.path.append('src')

import reversi

import numpy as np
import pylab as plt
import seaborn as sns

import time

#from tqdm import tqdm_notebook as tqdm
from tqdm import tqdm
sns.set()

import deepPlayer
import qFunction
importlib.reload(qFunction)

from memory import Memory
from qFunction import QFunction

import pickle

importlib.reload(deepPlayer)
#変更を加えたいファイルはreversiから出してreload(クラス名)をすればよい
```

# train

```python
episode_num = 10
memory_size = 64 * episode_num

np.set_printoptions(linewidth=100)
# memory_size = 100
PER = False
use_IS = False

alpha = 0.1 #0でrandom experience replayと等価

memory = Memory(size = memory_size)
eps_decrease = 1 / 2000

dqn_player = deepPlayer.DQNPlayer("どきゅん太郎", memory, eps_decrease = eps_decrease, alpha = alpha, PER=PER, use_IS=use_IS)

random_player = reversi.player.RandomPlayer("乱太郎")
```

## random

```python
random_train_num = 2000
random_train_num += 20

for i in tqdm(range(random_train_num)):
    if np.random.random() < 0.5:
        white = dqn_player
        black = random_player
    else:
        white = random_player
        black = dqn_player
    game = reversi.Reversi(white, black)
    game.main_loop(print_game = False)

    if dqn_player.memory.size <= dqn_player.memory.counter: #デバッグ用にメモリサイズ最大の時だけに
        dqn_player.q_function.train()
```

```python
scale = 5
plt.figure(figsize=(6*scale, 4*scale))
plt.hist(dqn_player.q_function.ind_log, bins=320)
plt.show()
```

```python
wining_DQN = np.array(dqn_player.record) == 1

final_win_rate = np.cumsum(wining_DQN)[len(np.cumsum(wining_DQN))-1]/len(wining_DQN)

title=[]
title.append("Win_rate (Battle with random(train))")
title.append("Final = {}, alpha={}".format(round(final_win_rate, 4), alpha))

plt.ylim(0, 1)
plt.title("\n".join(title))
plt.xlabel("episode")
plt.ylabel("win_rate")

plt.plot(np.cumsum(wining_DQN) / (np.arange(len(wining_DQN)) + 1))

#plt.plot(len(wining_DQN), final_win_rate, marker=".", markersize=10)

plt.show()
```

## own

```python
enemy_memory = Memory(size = memory_size)
enemy_dqn_player = deepPlayer.DQNPlayer("敵太郎", enemy_memory, eps_decrease = eps_decrease)
```

```python
train_num = 5000

for i in tqdm(range(train_num)):
    if np.random.random() < 0.5:
        white = dqn_player
        black = enemy_dqn_player
    else:
        white = enemy_dqn_player
        black = dqn_player
    game = reversi.Reversi(white, black)
    game.main_loop(print_game = False)


    dqn_player.q_function.train()
    enemy_dqn_player.q_function.train()
```

```python
wininig_DQN = np.array(dqn_player.record) == 1

plt.grid(True)
plt.ylim(0, 1)
plt.title("win_rate (Battle with same structure)")
plt.plot(np.cumsum(wininig_DQN) / (np.arange(len(wininig_DQN)) + 1))
plt.show()
```

# check win_rate(Battle with random)

```python
game_num = 2000

#train用情報退避
train_record = dqn_player.record
dqn_player.record = []

train_eps = dqn_player.eps
dqn_player.eps = 0 

train_memory = dqn_player.memory
dqn_player.memory = Memory(size=64)


for i in tqdm(range(game_num)):
    
    if np.random.random() < 0.5:
        white = random_player
        black = dqn_player
    else:
        white = dqn_player
        black = random_player
    game = reversi.Reversi(white, black)
    game.main_loop(print_game=False)
    


wining_DQN = np.array(dqn_player.record) == 1

plt.grid(True)
plt.ylim(0, 1)
final_win_rate = np.cumsum(wining_DQN)[len(np.cumsum(wining_DQN))-1]/len(wining_DQN)
title=[]
title.append("Win_rate (Battle with random(test))")
title.append("Final = {}, alpha={}".format(round(final_win_rate, 4), alpha))

plt.ylim(0, 1)
plt.title("\n".join(title))
plt.xlabel("episode")
plt.ylabel("win_rate")
plt.plot(np.cumsum(wining_DQN) / (np.arange(len(wining_DQN)) + 1))


#train用情報復帰
dqn_player.record = train_record
dqn_player.eps = train_eps
dqn_player.memory = train_memory
```

## check enemy win rate
同じ構造のネットワークと訓練した時は相手のネットワークの勝率も一応確認

```python
game_num = 2000

#train時レコードとepsilon退避
train_record = enemy_dqn_player.record
enemy_dqn_player.record = []

train_eps = enemy_dqn_player.eps
enemy_dqn_player.eps = 0 #greedy挙動のみ

train_memory = dqn_player.memory
dqn_player.memory = Memory(size=64)

for i in tqdm(range(game_num)):
    
    if np.random.random() < 0.5:
        white = random_player
        black = enemy_dqn_player
    else:
        white = enemy_dqn_player
        black = random_player
    game = reversi.Reversi(white, black)
    game.main_loop(print_game=False)
    


wining_DQN = np.array(enemy_dqn_player.record) == 1

plt.grid(True)
plt.ylim(0, 1)
final_win_rate = np.cumsum(wining_DQN)[len(np.cumsum(wining_DQN))-1]/len(wining_DQN)
plt.title("win_rate(Battle with random, final = {})".format(round(final_win_rate, 4)))
plt.xlabel("episode")
plt.ylabel("win_rate")
plt.plot(np.cumsum(wininig_DQN) / (np.arange(len(wininig_DQN)) + 1))


#recordを訓練用に復帰
enemy_dqn_player.record = train_record
enemy_dqn_player.eps = train_eps
enemy_dqn_player.memory = train_memory
```
